```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.spatial.distance import minkowski
df=pd.read_csv("extracted_features_charrec.csv")
num_columns = df.select_dtypes(include=['int','float']).columns.tolist()

# A1.  intraclass ,spread and interclass distances
def A1(data,num_columns):
    means=()
    spreads=()
    means=[np.mean(data[a], axis=0) for a in num_columns]
    spreads=[np.std(data[a], axis=0) for a in num_columns]
    interclass_distances = []
    for i in range(len(means)):
        for j in range(i + 1, len(means)):
            interclass_distances.append([[i,j], np.linalg.norm(means[i] - means[j])])
    return means, spreads,interclass_distances
# A2. Study feature density patterns

def A2(data,columns):
    bins=10
    plt.hist(data[columns], bins=bins)
    plt.xlabel(' Value')
    plt.ylabel('Frequency')
    plt.show()
    mean = np.mean(data[columns])
    variance = np.var(data[columns])
    return mean, variance

# A3.  Minkowski distance

def A3(vector1, vector2, r):
    distance = 0
    for i in range(len(vector1)):
        absolute_difference = abs(vector1[i] - vector2[i])
        distance += absolute_difference ** r

    return distance ** (1 / r)

# A4. Divide the dataset into train and test sets
def A4(X, y, test_size=0.3, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# A5. Train a kNN classifier
def A5(X_train, y_train, k=3):
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X_train, y_train)
    return knn_classifier




df.columns
```

```python
# A7. Use the predict function

def predict_class(knn_classifier, test_vector):
    predicted_class = knn_classifier.predict([test_vector])
    return predicted_class[0]

# A8. Compare kNN (k=3) with NN (k=1)

def A8_compare_knn_nn(X_train, X_test, y_train, y_test, k_values):
    accuracies = []
    for k in k_values:
        knn_classifier = KNeighborsClassifier(n_neighbors=k)
        knn_classifier.fit(X_train, y_train)
        accuracy = knn_classifier.score(X_test, y_test)
        accuracies.append(accuracy)
    return accuracies
```
